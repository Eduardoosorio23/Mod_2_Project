## Preprocessing and Train Test Split

```python
import pandas as pd
import numpy as np

kcc = pd.read_csv('D:\Flation\mod_2\Project2\Mod_2_Project/Mod_2 _Project/data_files/kc_cleaned.csv')
kcc.head()
```

```python
kcc.drop(['Unnamed: 0'], axis=1, inplace=True)
kcc.head()
```

```python
continuous = ['price', 'sqft_living', 'sqft_lot']
categoricals = ['floors', 'condition', 'waterfront', 'grade', 'zipcode', 'sale_month']
```

# Continuous Features:

```python
# Log transform and normalize
kcc_cont = kcc[continuous]

# log features
log_names = [f'{column}_log' for column in kcc_cont.columns]

kcc_log = np.log(kcc_cont)
kcc_log.columns = log_names

# normalize (subract mean and divide by std)
def normalize(feature):
    return (feature - feature.mean()) / feature.std()

kcc_log_norm = kcc_log.apply(normalize)
```

# Catagorical Features:

```python
kcc_ohe = pd.get_dummies(kcc[categoricals], columns=['floors','condition', 'waterfront', 'grade', 'zipcode', 'sale_month'])
kcc_ohe.head()
```

# Preprocessing:

```python
preprocessed_log = pd.concat([kcc_log_norm, kcc_ohe], axis=1)
preprocessed_log.head()
```

```python
X = preprocessed_log.drop('price_log', axis=1)
y = preprocessed_log['price_log']
```

# Train Test Split:

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
# Using a test size of 20%
```

# Fitting The Model

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()

linreg.fit(X_train, y_train)
y_hat_test = linreg.predict(X_test)
```

# Residuals and MSE

```python
from sklearn.metrics import mean_squared_error
test_residuals = y_hat_test - y_test

test_mse = mean_squared_error(y_test, y_hat_test)
test_mse
```

## Cross_validation:

```python
def kfolds(data, k):
    data = pd.DataFrame(data)
    num_observations = len(data)
    fold_size = num_observations//k
    leftovers = num_observations%k
    folds = []
    start_obs = 0
    for fold_n in range(1,k+1):
        if fold_n <= leftovers:
            fold =  data.iloc[start_obs : start_obs+fold_size+1] 
            folds.append(fold)
            start_obs +=  fold_size + 1
        else:
            fold =  data.iloc[start_obs : start_obs+fold_size] 
            folds.append(fold)
            start_obs +=  fold_size
            
    return folds 
# # Used the Kfolds fuction from one of the labs to increase the validation accuracy of our train test split 
```

```python
kcc_data = pd.concat([X.reset_index(drop=True), y], axis=1)
```

```python
kcc_folds = kfolds(kcc_data, 5)
```

```python
test_errs = []
train_errs = []
k=5

for n in range(k):
    # Split in train and test for the fold
    train = pd.concat([fold for i, fold in enumerate(kcc_folds) if i!=n])
    test = kcc_folds[n]
    # Fit a linear regression model
    linreg.fit(X_train, y_train)
    #Evaluate Train and Test Errors
    y_hat_train = linreg.predict(X_train)
    y_hat_test = linreg.predict(X_test)
    train_residuals = y_hat_train - y_train
    test_residuals = y_hat_test - y_test
    train_errs.append(np.mean(train_residuals.astype(float)**2))
    test_errs.append(np.mean(test_residuals.astype(float)**2))
print(train_errs)
print(test_errs)
```

```python
from sklearn.metrics import mean_squared_error, make_scorer
from sklearn.model_selection import cross_val_score

mse = make_scorer(mean_squared_error)

cv_10_results = cross_val_score(linreg, X, y, cv=10, scoring=mse)
```
